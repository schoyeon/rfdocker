```python
import pandas as pd
import os

# --- 1. 사용자 설정: 분석할 결과 파일 경로를 지정해주세요 ---
# 예시: 'outputs/0807-150-600/0807-150-600/mpnn_results.csv'
csv_file_path = "outputs/human-best-backbone/mpnn_results.csv" # <--- ★★★ 본인의 결과 CSV 파일 경로로 수정하세요 ★★★

# --- 2. 데이터 불러오기 ---
try:
    df = pd.read_csv(csv_file_path)

    # --- 결과를 저장할 단일 엑셀 파일 지정 ---
    output_excel_path = "analysis_summary.xlsx"
    with pd.ExcelWriter(output_excel_path, engine='openpyxl') as writer:

        # --- 3. 각 Backbone(design) 별로 통계 계산 ---
        metrics_to_analyze = ['mpnn', 'plddt', 'i_ptm', 'i_pae', 'rmsd']
        stats_by_design = df.groupby('design')[metrics_to_analyze].agg(['mean', 'std'])
        stats_by_design.columns = ['_'.join(col).strip() for col in stats_by_design.columns.values]

        # --- 4. [저장 1] 전체 뼈대 통계를 'Backbone_Stats' 시트에 저장 ---
        stats_by_design.to_excel(writer, sheet_name='Backbone_Stats')
        print("--- 뼈대(Backbone)별 성능 통계 (평균 및 표준편차) ---")
        print(stats_by_design)
        print("----------------------------------------------------")

        # --- 5. 가장 우수한 뼈대(Backbone) 찾기 ---
        best_backbones = stats_by_design.sort_values(by='rmsd_mean').head(10)
        print("\n--- 🏆 가장 안정적인 뼈대 (Top 10 Backbones by lowest mean RMSD) ---")
        print(best_backbones)
        print("----------------------------------------------------")

        # --- 6. 성공 기준 만족 후보 필터링 및 분석 ---
        successful_designs = df[(df['rmsd'] < 2) & (df['i_pae'] < 10)]

        print("\n\n------ 🌟 최상위 후보군 분석 (rmsd < 2 and i_pae < 10) ------")
        if successful_designs.empty:
            print("조건을 만족하는 최상위 후보를 찾지 못했습니다.")
        else:
            print(f"총 {len(successful_designs)}개의 최상위 후보가 발견되었습니다.")

            # --- 6-1. 성공 후보를 배출한 뼈대의 통계 계산 (hit_count, mean, std) ---
            metrics_for_success = ['rmsd', 'i_pae']
            successful_stats = successful_designs.groupby('design')[metrics_for_success].agg(['size', 'mean', 'std'])
            successful_stats.columns = ['_'.join(col).strip() for col in successful_stats.columns.values]
            successful_stats.rename(columns={'rmsd_size': 'hit_count'}, inplace=True)
            successful_stats = successful_stats.drop(columns=['i_pae_size'])
            
            # ******************************************************************* #
            # <<< ✨ 요청사항 반영: 개별 성공 목록에 뼈대 그룹의 통계(mean, std)를 병합합니다. ✨ >>>
            # 'design' ID를 기준으로, 개별 디자인 목록과 뼈대별 통계 정보를 합칩니다.
            successful_designs_with_stats = pd.merge(
                successful_designs,
                successful_stats,
                left_on='design',   # 개별 디자인 목록의 'design' 컬럼
                right_index=True,   # 뼈대 통계의 인덱스('design' ID)
                how='left'          # 왼쪽(개별 디자인 목록)을 기준으로 합치기
            )
            # ******************************************************************* #

            # --- [저장 2] 통계 정보가 추가된 성공 후보 전체 목록을 시트에 저장 ---
            successful_designs_with_stats.to_excel(writer, sheet_name='Successful_Designs_List', index=False)

            # --- 6-2. 0-hit을 포함한 전체 뼈대 통계 완성 ---
            all_designs_index = df['design'].unique()
            full_stats_by_backbone = successful_stats.reindex(all_designs_index).fillna(0)
            full_stats_by_backbone['hit_count'] = full_stats_by_backbone['hit_count'].astype(int)
            cols = ['hit_count'] + [col for col in full_stats_by_backbone if col != 'hit_count']
            full_stats_by_backbone = full_stats_by_backbone[cols]
            
            top_successful_backbones = full_stats_by_backbone.sort_values(by='hit_count', ascending=False).head(10)
            
            # --- [저장 3] 상위 10개 성공 뼈대 요약을 'Top_10_Backbones' 시트에 저장 ---
            top_successful_backbones.to_excel(writer, sheet_name='Top_10_Backbones')
            print("\n아래는 성공 후보를 가장 많이 배출한 상위 10개 뼈대(Backbone)의 요약입니다.")
            print(top_successful_backbones)

            # --- 6-3. Hit Count 별 뼈대 분포 분석 ---
            hit_count_distribution = full_stats_by_backbone['hit_count'].value_counts().sort_index(ascending=False)
            hit_count_distribution_df = hit_count_distribution.reset_index()
            hit_count_distribution_df.columns = ['hit_count', 'backbone_count']

            # --- [저장 4] Hit Count 분포를 'Hit_Count_Distribution' 시트에 저장 ---
            hit_count_distribution_df.to_excel(writer, sheet_name='Hit_Count_Distribution', index=False)
            print("\n\n------ 🎲 Hit Count 별 뼈대 분포 분석 ------")
            print("각 Hit Count에 해당하는 뼈대(Backbone)의 개수는 다음과 같습니다:")
            print(hit_count_distribution_df.to_string(index=False))

    # 최종 저장 완료 메시지
    print("\n================================================================")
    print(f"✅ 모든 분석 결과가 '{output_excel_path}' 파일 하나에 저장되었습니다.")
    print("================================================================")

except FileNotFoundError:
    print(f"⚠️ 파일을 찾을 수 없습니다: {csv_file_path}")
    print("위의 'csv_file_path' 변수를 본인의 결과 CSV 파일 경로에 맞게 수정했는지 확인해주세요.")
```
