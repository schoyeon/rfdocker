```python
# --- ÎùºÏù¥Î∏åÎü¨Î¶¨ import ---
import pandas as pd
import matplotlib.pyplot as plt
# ‚ú® pd.plottingÏóêÏÑú parallel_coordinatesÎ•º ÏßÅÏ†ë Î∂àÎü¨Ïò¨ ÏàòÎèÑ ÏûàÏäµÎãàÎã§.
from pandas.plotting import parallel_coordinates

# df Î≥ÄÏàòÏóêÎäî Ïã§Ï†ú Îç∞Ïù¥ÌÑ∞Í∞Ä Ïù¥ÎØ∏ Ï§ÄÎπÑÎêòÏñ¥ ÏûàÎã§Í≥† Í∞ÄÏ†ïÌï©ÎãàÎã§.

# --- ÌèâÌñâ Ï¢åÌëú ÌîåÎ°Ø ÏÉùÏÑ± ÏΩîÎìú ---

# 1. Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
df_parallel = df.copy()
df_parallel['neg_plddt'] = -df_parallel['plddt']
top_10_percent_threshold = df_parallel['rmsd'].quantile(0.1)
df_parallel['quality'] = 'Normal'
df_parallel.loc[df_parallel['rmsd'] <= top_10_percent_threshold, 'quality'] = 'BEST'

# 2. ÌîåÎ°Ø ÏÉùÏÑ±
plt.figure(figsize=(12, 7))

# <<< Ìï¥Í≤∞Ï±Ö: Ìï®Ïàò ÏïûÏóê 'pd.plotting.'ÏùÑ Î∂ôÏó¨Ï£ºÍ±∞ÎÇò, ÏúÑÏóêÏÑúÏ≤òÎüº from ... import ... Íµ¨Î¨∏ÏùÑ ÏÇ¨Ïö©Ìï©ÎãàÎã§.
parallel_coordinates(
    df_parallel[['rmsd', 'neg_plddt', 'i_pae', 'quality']],
    'quality',
    color=['#cccccc', '#007acc'],
    linewidth=2
)

# 3. Ï†úÎ™©Í≥º Ï∂ï ÎùºÎ≤® Ï∂îÍ∞Ä
plt.title('Parallel Coordinates Plot of Design Scores', fontsize=16)
plt.xlabel('Metrics')
plt.ylabel('Value (pLDDT is negative)')
plt.grid(True)
plt.legend(title='Design Quality')

# 4. ÌîåÎ°ØÏùÑ ÌôîÎ©¥Ïóê Î≥¥Ïó¨Ï§çÎãàÎã§.
plt.show()
```
